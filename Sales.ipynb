# Exploratory Data Analysis

```python
plt.figure(figsize=(15,5))
sns.countplot(x="Item Type", data=df)
#to show no of orders placed for each item
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x="Units Sold",y="Item Type",data=df)
```

```python
plt.figure(figsize=(60,40))
sns.catplot(x="Units Sold", col="Region",data=df)
plt.show()
# the sale in Sub-Saharan Africa and Europe is good but the sale in North America is very low
```

```python
sns.catplot(x="Unit Price", y="Item Type", col="Region", data=df)
#Unit Price is constant everywhere
```

```python
sns.catplot(x="Units Sold", y="Item Type", col="Region", hue='Sales Channel',data=df)
```

```python
plt.figure(figsize=(15,15))
sns.jointplot(y='Item Type', x='Total Profit',data=df)
plt.show()
#Office supplies Costmetics and Household item types are top 3 profitable items
```

```python
sns.relplot(x="Total Profit", y="Units Sold", hue="Item Type", data=df);
#household and cosmetics are almost Equally Profitable followed by office supplies
```

```python
sns.displot(df, x="Total Profit", hue="Region", kind="kde", fill=False)
```

```python
sns.displot(df, x="Total Profit", hue="Region", kind="ecdf")
#Proportion of profit is equal in almost every region
```

```python
#top 20 countries which made highest profit.
info = pd.DataFrame(df['Total Profit'].sort_values(ascending = False))
info['Country'] = df['Country']
data = list(map(str,(info['Country'])))


x = list(data[:20])
y = list(info['Total Profit'][:20])
ax = sns.barplot(x=y,y=x)
```

```python
#Least  20  profitable countries .
info = pd.DataFrame(df['Total Profit'].sort_values(ascending = True))
info['Country'] = df['Country']
data = list(map(str,(info['Country'])))


x = list(data[:20])
y = list(info['Total Profit'][:20])
ax = sns.barplot(x=y,y=x)
#these are the countries company needs to work on
```

```python
sns.displot(df, x="Total Profit", hue="Sales Channel",col='Region', kind="kde")
#Proportion of profit is equal in both channels in each region
```

```python
df["Year"] = df["Order Date"].str.replace("\\", "/")
df["Year"] = pd.to_datetime(df["Year"])
info['Year']=pd.DataFrame(df["Year"].dt.year)
info["Profit"]=df["Total Profit"]
sns.lineplot(x='Year',y='Profit',data=info)
# profit is decreasing since last two years
```

```python
df["Year"] = df["Order Date"].str.replace("\\", "/")
df["Year"] = pd.to_datetime(df["Year"])
df['Year']=df["Year"].dt.year
plot = sns.FacetGrid(df, col="Region")
plot.map(plt.bar, "Year","Total Profit")
plt.show()
#there is a huge decrease in profit in North America in 2017
#also the profit is almost constant over all these years so company needs to look into that also
```

```python
yearly_average = info[['Year','Profit']].groupby('Year', as_index=False).mean()
yearly_average[:10]

yearly_average[-20:].plot(x='Year', y='Profit', figsize=(10,5), grid=True)
```
